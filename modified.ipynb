```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo

# Fetch dataset
dataset = fetch_ucirepo(id=967)
X = dataset.data.features
y = dataset.data.targets

# Print dataset metadata and variable information
print(dataset.metadata)
print(dataset.variables)

# Convert y to binary labels (phishing or not)
y = y.replace(-1, 0)

# Select only integer columns from X
X = X.select_dtypes(include=[int, float])

# Normalize the integer features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Ensure the target tensors are 1D
y_train = y_train.squeeze()
y_test = y_test.squeeze()

# Convert the data to PyTorch tensors
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_train.values, dtype=torch.long)
y_test = torch.tensor(y_test.values, dtype=torch.long)

# Define the DenseNet block
class DenseBlock(nn.Module):
    def __init__(self, in_features, growth_rate, num_layers):
        super(DenseBlock, self).__init__()
        self.layers = nn.ModuleList()
        for i in range(num_layers):
            layer = nn.Sequential(
                nn.Linear(in_features + i * growth_rate, growth_rate),
                nn.BatchNorm1d(growth_rate),
                nn.ReLU(inplace=True)
            )
            self.layers.append(layer)

    def forward(self, x):
        features = [x]
        for layer in self.layers:
            new_features = layer(torch.cat(features, dim=1))
            features.append(new_features)
        return torch.cat(features, dim=1)

# Define the DenseNet model
class DenseNet(nn.Module):
    def __init__(self, num_features, num_classes, growth_rate=32, num_layers_per_block=4):
        super(DenseNet, self).__init__()
        self.input_layer = nn.Linear(num_features, growth_rate)
        self.dense_block1 = DenseBlock(growth_rate, growth_rate, num_layers_per_block)
        self.transition_layer1 = nn.Linear(growth_rate + num_layers_per_block * growth_rate, growth_rate)
        self.dense_block2 = DenseBlock(growth_rate, growth_rate, num_layers_per_block)
        self.transition_layer2 = nn.Linear(growth_rate + num_layers_per_block * growth_rate, growth_rate)
        self.output_layer = nn.Linear(growth_rate, num_classes)

    def forward(self, x):
        x = F.relu(self.input_layer(x))
        x = self.dense_block1(x)
        x = F.relu(self.transition_layer1(x))
        x = self.dense_block2(x)
        x = F.relu(self.transition_layer2(x))
        x = self.output_layer(x)
        return F.log_softmax(x, dim=1)

# Model parameters
num_features = X_train.shape[1]
num_classes = 2  # Phishing or not

# Instantiate the model
model = DenseNet(num_features, num_classes)

# Define optimizer and loss function
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
criterion = torch.nn.CrossEntropyLoss()

# Training function
def train(model, optimizer, criterion):
    model.train()
    optimizer.zero_grad()
    output = model(X_train)
    loss = criterion(output, y_train)
    loss.backward()
    optimizer.step()
    return loss.item()

# Evaluate the model
def evaluate(model, data, labels):
    model.eval()
    with torch.no_grad():
        output = model(data)
        pred = output.argmax(dim=1)
        accuracy = accuracy_score(labels.cpu(), pred.cpu())
        precision = precision_score(labels.cpu(), pred.cpu())
        recall = recall_score(labels.cpu(), pred.cpu())
        f1 = f1_score(labels.cpu(), pred.cpu())
    return accuracy, precision, recall, f1

# Store metrics
losses = []
train_accuracies, train_precisions, train_recalls, train_f1s = [], [], [], []
test_accuracies, test_precisions, test_recalls, test_f1s = [], [], [], []

# Training loop
for epoch in range(1, 51):
    loss = train(model, optimizer, criterion)
    losses.append(loss)
    if epoch % 10 == 0:
        train_accuracy, train_precision, train_recall, train_f1 = evaluate(model, X_train, y_train)
        test_accuracy, test_precision, test_recall, test_f1 = evaluate(model, X_test, y_test)
        train_accuracies.append(train_accuracy)
        train_precisions.append(train_precision)
        train_recalls.append(train_recall)
        train_f1s.append(train_f1)
        test_accuracies.append(test_accuracy)
        test_precisions.append(test_precision)
        test_recalls.append(test_recall)
        test_f1s.append(test_f1)
        print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}')
        print(f'Train - Accuracy: {train_accuracy:.4f}, Precision: {train_precision:.4f}, Recall: {train_recall:.4f}, F1: {train_f1:.4f}')
        print(f'Test - Accuracy: {test_accuracy:.4f}, Precision: {test_precision:.4f}, Recall: {test_recall:.4f}, F1: {test_f1:.4f}')

# Plotting training loss over epochs
plt.figure(figsize=(12, 5))
plt.plot(range(1, 51), losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.show()

# Plotting evaluation metrics
epochs = range(10, 51, 10)

plt.figure(figsize=(12, 5))
plt.plot(epochs, train_accuracies, label='Train Accuracy')
plt.plot(epochs, test_accuracies, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy over Epochs')
plt.legend()
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(epochs, train_precisions, label='Train Precision')
plt.plot(epochs, test_precisions, label='Test Precision')
plt.xlabel('Epoch')
plt.ylabel('Precision')
plt.title('Precision over Epochs')
plt.legend()
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(epochs, train_recalls, label='Train Recall')
plt.plot(epochs, test_recalls, label='Test Recall')
plt.xlabel('Epoch')
plt.ylabel('Recall')
plt.title('Recall over Epochs')
plt.legend()
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(epochs, train_f1s, label='Train F1 Score')
plt.plot(epochs, test_f1s, label='Test F1 Score')
plt.xlabel('Epoch')
plt.ylabel('F1 Score')
plt.title('F1 Score over Epochs')
plt.legend()
plt.show()
```

```python
import re
import requests
from urllib.parse import urlparse
import tldextract
from bs4 import BeautifulSoup
from difflib import SequenceMatcher

def extract_features(url):
    features = {}

    # Helper function to check if a string is an IP address
    def is_ip(address):
        try:
            return all(0 <= int(num) < 256 for num in address.split('.'))
        except ValueError:
            return False

    # Helper function to calculate similarity between two strings
    def similarity(a, b):
        return SequenceMatcher(None, a, b).ratio()

    # Normalize URL to start with http:// if missing
    if not url.startswith(('http://', 'https://')):
        url = 'http://' + url

    # URL parsing
    parsed_url = urlparse(url)
    extracted = tldextract.extract(url)

    # Lexical features
    features['URLLength'] = len(url)
    features['DomainLength'] = len(parsed_url.hostname) if parsed_url.hostname else 0
    features['IsDomainIP'] = 1 if is_ip(parsed_url.hostname) else 0
    features['TLDLength'] = len(extracted.suffix)
    features['NoOfSubDomain'] = len(extracted.subdomain.split('.')) if extracted.subdomain else 0
    features['NoOfObfuscatedChar'] = sum(c in '%25' for c in url)  # Percent-encoded characters
    features['NoOfLettersInURL'] = sum(c.isalpha() for c in url)
    features['LetterRatioInURL'] = features['NoOfLettersInURL'] / features['URLLength']
    features['NoOfDegitsInURL'] = sum(c.isdigit() for c in url)
    features['DegitRatioInURL'] = features['NoOfDegitsInURL'] / features['URLLength']
    features['NoOfEqualsInURL'] = url.count('=')
    features['NoOfQMarkInURL'] = url.count('?')
    features['NoOfAmpersandInURL'] = url.count('&')
    features['NoOfOtherSpecialCharsInURL'] = len(re.findall(r'[^a-zA-Z0-9]', url))
    features['SpacialCharRatioInURL'] = features['NoOfOtherSpecialCharsInURL'] / features['URLLength']
    features['IsHTTPS'] = 1 if parsed_url.scheme == 'https' else 0

    # Initialize complex features with default values
    features.update({
        'URLSimilarityIndex': 0,
        'CharContinuationRate': 0,
        'TLDLegitimateProb': 0.0,
        'URLCharProb': 0.0,
        'HasObfuscation': 0,
        'ObfuscationRatio': 0,
        'DomainTitleMatchScore': 0,
        'URLTitleMatchScore': 0,
        'Robots': 0,
        'IsResponsive': 0,
        'NoOfURLRedirect': 0,
        'NoOfSelfRedirect': 0,
        'HasDescription': 0,
        'Bank': 0,
        'Pay': 0,
        'Crypto': 0
    })

    try:
        response = requests.get(url, timeout=5)
        if response.status_code == 200:
            page_content = response.content
            soup = BeautifulSoup(page_content, 'html.parser')
            
            # Content-based features
            features['LineOfCode'] = page_content.decode().count('\n')
            features['LargestLineLength'] = max(len(line) for line in page_content.decode().split('\n'))
            features['HasTitle'] = 1 if soup.title else 0
            features['HasFavicon'] = 1 if soup.find('link', rel='icon') else 0
            features['NoOfPopup'] = len(soup.find_all('popup'))
            features['NoOfiFrame'] = len(soup.find_all('iframe'))
            features['HasExternalFormSubmit'] = 1 if any(form.get('action') and urlparse(form.get('action')).hostname != parsed_url.hostname for form in soup.find_all('form')) else 0
            features['HasSocialNet'] = 1 if any(a['href'].startswith(('https://www.facebook.com', 'https://twitter.com', 'https://www.linkedin.com')) for a in soup.find_all('a', href=True)) else 0
            features['HasSubmitButton'] = 1 if soup.find('button', type='submit') else 0
            features['HasHiddenFields'] = 1 if any(input_['type'] == 'hidden' for input_ in soup.find_all('input', type=True)) else 0
            features['HasPasswordField'] = 1 if soup.find('input', type='password') else 0
            features['HasCopyrightInfo'] = 1 if 'Â©' in soup.text else 0
            features['NoOfImage'] = len(soup.find_all('img'))
            features['NoOfCSS'] = len(soup.find_all('link', rel='stylesheet'))
            features['NoOfJS'] = len(soup.find_all('script'))
            features['NoOfSelfRef'] = sum(1 for a in soup.find_all('a', href=True) if a['href'].startswith('#'))
            features['NoOfEmptyRef'] = sum(1 for a in soup.find_all('a', href=True) if a['href'] == '')
            features['NoOfExternalRef'] = sum(1 for a in soup.find_all('a', href=True) if urlparse(a['href']).hostname not in [None, parsed_url.hostname])

            # URLSimilarityIndex
            domain_parts = extracted.domain.split('.')
            if len(domain_parts) > 1:
                features['URLSimilarityIndex'] = similarity(domain_parts[0], domain_parts[1])

            # CharContinuationRate
            features['CharContinuationRate'] = sum(1 for i in range(1, len(url)) if url[i] == url[i-1]) / len(url)
            
            # TLDLegitimateProb and URLCharProb (assumed probability based on known TLDs and char ratios)
            legitimate_tlds = ['com', 'org', 'net', 'edu', 'gov']
            features['TLDLegitimateProb'] = 1.0 if extracted.suffix in legitimate_tlds else 0.1
            features['URLCharProb'] = features['LetterRatioInURL']

            # HasObfuscation and ObfuscationRatio
            obfuscation_chars = sum(c in '%25' for c in url)
            features['HasObfuscation'] = 1 if obfuscation_chars > 0 else 0
            features['ObfuscationRatio'] = obfuscation_chars / features['URLLength']
            
            # DomainTitleMatchScore and URLTitleMatchScore
            if soup.title:
                title = soup.title.string
                features['DomainTitleMatchScore'] = similarity(extracted.domain, title)
                features['URLTitleMatchScore'] = similarity(url, title)

            # Robots
            robots_txt = requests.get(parsed_url.scheme + '://' + parsed_url.hostname + '/robots.txt', timeout=5)
            features['Robots'] = 1 if robots_txt.status_code == 200 else 0

            # IsResponsive (basic check if the page is responsive)
            meta_viewport = soup.find('meta', attrs={'name': 'viewport'})
            features['IsResponsive'] = 1 if meta_viewport and 'width=device-width' in meta_viewport.get('content', '') else 0

            # NoOfURLRedirect and NoOfSelfRedirect (assuming a simple count of meta-refresh tags)
            meta_refresh = soup.find('meta', attrs={'http-equiv': 'refresh'})
            if meta_refresh:
                content = meta_refresh.get('content', '')
                url_redirects = re.findall(r'url=(.*)', content, re.IGNORECASE)
                features['NoOfURLRedirect'] = len(url_redirects)
                features['NoOfSelfRedirect'] = sum(1 for u in url_redirects if urlparse(u).hostname == parsed_url.hostname)

            # HasDescription
            meta_description = soup.find('meta', attrs={'name': 'description'})
            features['HasDescription'] = 1 if meta_description else 0

            # Bank, Pay, Crypto (checking for specific keywords in the content)
            page_text = soup.get_text().lower()
            features['Bank'] = 1 if 'bank' in page_text else 0
            features['Pay'] = 1 if 'pay' in page_text else 0
            features['Crypto'] = 1 if any(keyword in page_text for keyword in ['crypto', 'bitcoin', 'ethereum']) else 0

        else:
            # If the page is not accessible, assign zero or default values
            for feature in ['LineOfCode', 'LargestLineLength', 'HasTitle', 'HasFavicon', 'NoOfPopup', 'NoOfiFrame',
                            'HasExternalFormSubmit', 'HasSocialNet', 'HasSubmitButton', 'HasHiddenFields',
                            'HasPasswordField', 'HasCopyrightInfo', 'NoOfImage', 'NoOfCSS', 'NoOfJS',
                            'NoOfSelfRef', 'NoOfEmptyRef', 'NoOfExternalRef']:
                features[feature] = 0
    except Exception as e:
        # If any exception occurs, assign zero or default values
        for feature in ['LineOfCode', 'LargestLineLength', 'HasTitle', 'HasFavicon', 'NoOfPopup', 'NoOfiFrame',
                        'HasExternalFormSubmit', 'HasSocialNet', 'HasSubmitButton', 'HasHiddenFields',
                        'HasPasswordField', 'HasCopyrightInfo', 'NoOfImage', 'NoOfCSS', 'NoOfJS',
                        'NoOfSelfRef', 'NoOfEmptyRef', 'NoOfExternalRef']:
            features[feature] = 0

    return features

# Example usage
url = 'http://www.kaggle.com'
feature_names = extract_features(url)
print(features)
```
